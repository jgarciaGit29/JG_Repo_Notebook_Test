## Introduction

In this notebook, we will use the Iris Flower Dataset, which is a classic dataset in the field of machine learning. This dataset contains measurements of iris flowers from three different species: setosa, versicolor, and virginica. The dataset includes the following features:
- Sepal length
- Sepal width
- Petal length
- Petal width

We will load this dataset, perform some basic data analysis, and create visualizations to better understand the data.


```python
import pandas as pd
from sklearn.datasets import load_iris

# Load the iris dataset from sklearn
iris = load_iris()
iris_df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
iris_df['species'] = iris.target

# Display the first few rows of the dataset
iris_df.head()
```

### Basic Information

Let's start by displaying some basic information about the dataset, including its shape, data types, and summary statistics.


```python
# Display the shape of the dataset
print(f'The dataset contains {iris_df.shape[0]} rows and {iris_df.shape[1]} columns.')

# Display data types and non-null counts
print('\nData Types and Non-Null Counts:')
print(iris_df.info())

# Display summary statistics
print('\nSummary Statistics:')
print(iris_df.describe())
```

### Missing Values

Next, we will check for any missing values in the dataset.


```python
# Check for missing values
missing_values = iris_df.isnull().sum()
print('\nMissing Values:')
print(missing_values)
```

### Data Visualization

Now, let's create some visualizations to understand the distribution of the features and the relationships between them.


```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set the style of the visualizations
sns.set_theme(style="whitegrid")

# Pairplot to visualize relationships between features
sns.pairplot(iris_df, hue="species", palette="viridis")
plt.suptitle('Pairplot of Iris Dataset', y=1.02)
plt.show()

# Boxplot to visualize the distribution of features
plt.figure(figsize=(12, 6))
sns.boxplot(data=iris_df, orient="h", palette="Set2")
plt.title('Boxplot of Iris Dataset Features')
plt.show()
```

### Pairplot

A pairplot is a great way to visualize relationships between different features in the dataset. It creates a matrix of scatter plots for each pair of features. This helps us understand how the features are correlated and how the species are distributed in the feature space. The diagonal shows the distribution of each feature. By color-coding the species, we can see how different species are separated based on the features.


### Boxplot

A boxplot provides a summary of the distribution of individual features. It displays the median, quartiles, and potential outliers for each feature. This visualization helps us understand the spread and skewness of the data. By comparing the boxplots of different features, we can identify any significant differences in their distributions. Boxplots are particularly useful for spotting outliers and comparing the variability of different features.
